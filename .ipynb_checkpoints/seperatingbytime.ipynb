```python
import pandas as pd

# Load JSON file into DataFrame
file_path = '/Users/uyakut/Desktop/CaDence-old/simeventdata/listen_events.json'  # Replace with your JSON file path
df = pd.read_json(file_path, lines=True)

# Convert 'ts' column from milliseconds to seconds
df['ts_seconds'] = df['ts'] // 1000

# Sort the DataFrame by timestamp ('ts_seconds')
df = df.sort_values('ts_seconds')

# Create a 10-second interval grouping based on sorted timestamps
df['group'] = ((df['ts_seconds'] - df['ts_seconds'].min()) // 60)

# Get the number of groups (intervals)
num_groups = df['group'].nunique()

# Set the maximum number of files
max_files = 5000

# Calculate how many groups should be in each file
groups_per_file = num_groups // max_files + 1 if num_groups > max_files else 1

# Split the DataFrame into groups based on the grouping column
groups = df.groupby('group')

# Save the groups in batches, limiting the number of files to max_files
current_file = 1
group_list = []

for idx, (group, group_data) in enumerate(groups, start=1):
    group_list.append(group_data)
    
    # If we've collected enough groups for one file, save it and reset
    if len(group_list) >= groups_per_file:
        output_file = f'group_{current_file}.json'
        pd.concat(group_list).drop(columns='group').to_json(output_file, orient='records', lines=True)
        print(f"Saved: {output_file}")
        current_file += 1
        group_list = []  # Reset the list for the next file
    
    # Stop after 10,000 files
    if current_file > max_files:
        break

```
